```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
# Preparation des donn√©es
df_diabetes = pd.read_csv('diabetes.csv')
```

```python
df_diabetes.info()
```

```python
df_diabetes.head()
```

```python
df_diabetes.describe()
```

```python
# √âtape nettoyage des donn√©es
```

```python
# 1. V√©rifier les valeurs manquantes
df_diabetes.isnull().sum() # Pas de valeurs manquantes
```

```python
# 2. Supprimer ou g√©rer les doublons
df_diabetes.duplicated().sum() # Pas de doublons
```

```python
(df_diabetes == 0).sum()
```

```python
# Liste des colonnes ou la valeur 0 n'est pas biologiquement plausible
cols_with_zeros = [ 'Glucose', 'BloodPressure', 'BMI', 'SkinThickness', 'Insulin' ]
# Je remplace les 0 par la m√©diane pour chaque colonne concern√©e
for col in cols_with_zeros :
    # Je calcule la m√©diane uniquement sur les colonnes differents de 0
    median = df_diabetes[col][df_diabetes[col] != 0 ].median()
    # Je remplace les 0 par cette m√©diane
    df_diabetes[col] = df_diabetes[col].replace(0, median)
```

```python
print((df_diabetes[cols_with_zeros] == 0).sum())
```

```python
df_diabetes.describe()
```

```python
df_diabetes['Outcome'] = df_diabetes['Outcome'].astype(bool)
```

```python
df_diabetes.describe()
```

```python
df_diabetes.duplicated().sum()
```

```python
# Normalisation avec StandardScaler
numeric_columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_diabetes[numeric_columns])

df_scaled = pd.DataFrame(X_scaled, columns=numeric_columns)


df_scaled['Outcome'] = df_diabetes['Outcome'].values

df_scaled.head()
```

```python
# 2. Nettoyage : supprimer les lignes o√π la cible est NaN
print(df_scaled['Outcome'].value_counts())
```

```python
# Analyse exploratoire des donn√©es(EDA)
```

```python
# Visualiser chaque distribution avec un histogramme
for col in numeric_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df_scaled[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution de {col} (normalis√©e)')
    plt.xlabel(col)
    plt.ylabel('Fr√©quence')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# un graphique de la colonne cible
sns.countplot(x='Outcome', data=df_scaled, hue='Outcome', palette='pastel', legend=False)
plt.title('R√©partition des patients diab√©tiques vs non diab√©tiques')
plt.xlabel('Diab√©tique (False = non, True = oui)')
plt.ylabel('Nombre de patients')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(9, 4))
corr = df_scaled.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matrice de corr√©lation')
plt.show()
```

```python
corr_target = df_scaled.corr()['Outcome'].sort_values(ascending=False)
print(corr_target)
```

```python
# Scatter plots avec la cible 
for col in numeric_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='Outcome', y=col, data=df_scaled, hue='Outcome', palette='Set2', legend=False)
    plt.title(f'{col} en fonction du diab√®te')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
X = df_scaled.drop(['Outcome'], axis=1)
y = df_scaled['Outcome']
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)
```

```python
# S√©paration des donn√©es en ensembles d'entra√Ænement et de test
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)
```

```python
# Cr√©ation du mod√®le de r√©gression logistique
model = LogisticRegression(class_weight='balanced')
# Entra√Ænement du mod√®le
model.fit(X_train, y_train)
```

```python
import joblib

# Sauvegarder les objets entra√Æn√©s
joblib.dump(scaler, 'scaler_Diab√®te.pkl')
joblib.dump(model, 'logistic_model_Diab√®te.pkl')

print("‚úÖ Sauvegarde r√©ussie.")
```

```python
# Pr√©diction sur l'ensemble de test
y_pred = model.predict(X_test)

# √âvaluation du mod√®le
print(f'Accuracy: {accuracy_score(y_test, y_pred):.4f}')
print('Classification Report:')
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Matrice de confusion
cm = confusion_matrix(y_test, y_pred)

# Affichage
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Non diab√©tique", "Diab√©tique"])
disp.plot(cmap='Blues', values_format='d')
plt.title('Matrice de confusion')
plt.grid(False)
plt.show()
```

```python
# ---------------------------------------------------------
# Pr√©diction d'un nouveau patient ‚Äì Instructions
# ---------------------------------------------------------
# Entrez ici les informations m√©dicales d‚Äôun patient (valeurs brutes, non normalis√©es).
#
# ‚ö†Ô∏è IMPORTANT :
#    - Les valeurs doivent rester dans les plages observ√©es du dataset d'origine (avant normalisation).
#    - Le code applique automatiquement la normalisation avec le m√™me scaler que pour l'entra√Ænement.
#    - Assurez-vous que les donn√©es saisies sont r√©alistes (ex : pas de z√©ros anormaux).
#
# Plages typiques observ√©es dans le dataset (valeurs recommand√©es) :
#    - Pregnancies                 : 0 ‚Äì 17
#    - Glucose                     : 44 ‚Äì 199
#    - BloodPressure               : 24 ‚Äì 122
#    - SkinThickness               : 7 ‚Äì 99
#    - Insulin                     : 14 ‚Äì 846
#    - BMI                         : 18.2 ‚Äì 67.1
#    - DiabetesPedigreeFunction    : 0.078 ‚Äì 2.42
#    - Age                         : 21 ‚Äì 81
#
# Exemple :
#    manual_input = [[2, 130, 70, 28, 120, 31.5, 0.4, 30]]
#
# Le mod√®le pr√©dira ensuite si le patient est atteint de diab√®te (1) ou non (0).
# ---------------------------------------------------------
```

```python
# Prediction sur un nouveau patient
manual_input = [[
    10, 130, 70, 28, 120, 31.5, 0.4, 30
]]
manual_input_df = pd.DataFrame(manual_input, columns=numeric_columns)

manual_input_scaled = scaler.transform(manual_input_df)
manual_input_scaled_df = pd.DataFrame(manual_input_scaled, columns=numeric_columns)
```

```python
# Pr√©diction brute (0 ou 1)
prediction = model.predict(manual_input_scaled_df)[0]

# Probabilit√© que la patiente soit diab√©tique
proba = model.predict_proba(manual_input_scaled_df)[0][1]

print(f"R√©sultat pr√©diction : {'Diab√©tique' if prediction else 'Non diab√©tique'}")
print(f"Probabilit√© : {proba:.2%}")
```
